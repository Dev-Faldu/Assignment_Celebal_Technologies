```python

def set_plot_theme():
    """Set consistent plotting theme across all visualizations"""
    plt.rcParams['figure.figsize'] = (12, 8)
    plt.rcParams['font.size'] = 12
    plt.rcParams['axes.titlesize'] = 14
    plt.rcParams['axes.labelsize'] = 12
    plt.rcParams['xtick.labelsize'] = 10
    plt.rcParams['ytick.labelsize'] = 10
    plt.rcParams['legend.fontsize'] = 10

def create_summary_stats(df, column):
    """Generate comprehensive summary statistics for a column"""
    stats_dict = {
        'Count': df[column].count(),
        'Missing': df[column].isnull().sum(),
        'Missing %': (df[column].isnull().sum() / len(df)) * 100,
        'Unique': df[column].nunique(),
        'Data Type': df[column].dtype
    }

    if df[column].dtype in ['int64', 'float64']:
        stats_dict.update({
            'Mean': df[column].mean(),
            'Median': df[column].median(),
            'Std': df[column].std(),
            'Min': df[column].min(),
            'Max': df[column].max(),
            'Skewness': df[column].skew(),
            'Kurtosis': df[column].kurtosis()
        })

    return pd.Series(stats_dict)

def detect_outliers_iqr(df, column):
    """Detect outliers using IQR method"""
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

def cramers_v_stat(x, y):
    """Calculate Cram√©r's V statistic for categorical association"""
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))

set_plot_theme()
```

```python


def classify_features(df):
    """Classify features into different types"""
    numerical_features = df.select_dtypes(include=[np.number]).columns.tolist()
    categorical_features = df.select_dtypes(include=['object']).columns.tolist()

    # Further classify numerical features
    discrete_features = []
    continuous_features = []

    for col in numerical_features:
        if df[col].nunique() < 20:  # Heuristic for discrete vs continuous
            discrete_features.append(col)
        else:
            continuous_features.append(col)

    return {
        'numerical': numerical_features,
        'categorical': categorical_features,
        'discrete': discrete_features,
        'continuous': continuous_features
    }

feature_types = classify_features(df)

print("="*60)
print("FEATURE CLASSIFICATION")
print("="*60)
for feature_type, features in feature_types.items():
    print(f"{feature_type.upper()}: {features}")

# Visualize feature distribution
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Feature type distribution
feature_counts = [len(feature_types['numerical']), len(feature_types['categorical'])]
feature_labels = ['Numerical', 'Categorical']

axes[0].pie(feature_counts, labels=feature_labels, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
axes[0].set_title('Distribution of Feature Types', fontsize=14, fontweight='bold')

# Data type distribution
dtype_counts = df.dtypes.value_counts()
axes[1].bar(range(len(dtype_counts)), dtype_counts.values, color='lightgreen')
axes[1].set_xticks(range(len(dtype_counts)))
axes[1].set_xticklabels(dtype_counts.index, rotation=45)
axes[1].set_title('Distribution of Data Types', fontsize=14, fontweight='bold')
axes[1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python

def advanced_numerical_analysis(df, columns):
    """Perform advanced statistical analysis on numerical columns"""
    results = []

    for col in columns:
        if col in df.columns:
            data = df[col].dropna()

            # Normality tests
            shapiro_stat, shapiro_p = stats.shapiro(data.sample(min(5000, len(data))))

            # Distribution fitting
            distributions = ['norm', 'lognorm', 'expon', 'gamma']
            best_dist = None
            best_p = 0

            for dist_name in distributions:
                try:
                    dist = getattr(stats, dist_name)
                    params = dist.fit(data)
                    ks_stat, ks_p = stats.kstest(data, lambda x: dist.cdf(x, *params))
                    if ks_p > best_p:
                        best_p = ks_p
                        best_dist = dist_name
                except:
                    continue

            results.append({
                'Column': col,
                'Shapiro_Stat': shapiro_stat,
                'Shapiro_p_value': shapiro_p,
                'Is_Normal': shapiro_p > 0.05,
                'Best_Distribution': best_dist,
                'Best_Fit_p_value': best_p,
                'Skewness': data.skew(),
                'Kurtosis': data.kurtosis(),
                'CV': data.std() / data.mean() if data.mean() != 0 else np.inf
            })

    return pd.DataFrame(results)

# Perform advanced analysis on numerical columns
if numerical_cols:
    advanced_stats = advanced_numerical_analysis(df, numerical_cols)
    print("="*60)
    print("ADVANCED STATISTICAL ANALYSIS")
    print("="*60)
    print(advanced_stats)
```

```python

def comprehensive_outlier_analysis(df, columns):
    """Perform comprehensive outlier analysis"""
    outlier_summary = []

    fig, axes = plt.subplots(len(columns), 3, figsize=(18, 6*len(columns)))
    if len(columns) == 1:
        axes = axes.reshape(1, -1)

    for i, col in enumerate(columns):
        if col in df.columns:
            data = df[col].dropna()

            # IQR method
            Q1 = data.quantile(0.25)
            Q3 = data.quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            iqr_outliers = data[(data < lower_bound) | (data > upper_bound)]

            # Z-score method
            z_scores = np.abs(stats.zscore(data))
            z_outliers = data[z_scores > 3]

            # Modified Z-score method
            median = data.median()
            mad = np.median(np.abs(data - median))
            modified_z_scores = 0.6745 * (data - median) / mad
            modified_z_outliers = data[np.abs(modified_z_scores) > 3.5]

            outlier_summary.append({
                'Column': col,
                'IQR_Outliers': len(iqr_outliers),
                'IQR_Percentage': len(iqr_outliers) / len(data) * 100,
                'Z_Score_Outliers': len(z_outliers),
                'Z_Score_Percentage': len(z_outliers) / len(data) * 100,
                'Modified_Z_Outliers': len(modified_z_outliers),
                'Modified_Z_Percentage': len(modified_z_outliers) / len(data) * 100
            })

            # Visualizations
            # Boxplot
            axes[i, 0].boxplot(data, vert=True)
            axes[i, 0].set_title(f'{col} - Boxplot', fontweight='bold')
            axes[i, 0].set_ylabel(col)

            # Histogram with outlier boundaries
            axes[i, 1].hist(data, bins=30, alpha=0.7, color='skyblue', edgecolor='black')
            axes[i, 1].axvline(lower_bound, color='red', linestyle='--', label=f'IQR Lower: {lower_bound:.2f}')
            axes[i, 1].axvline(upper_bound, color='red', linestyle='--', label=f'IQR Upper: {upper_bound:.2f}')
            axes[i, 1].set_title(f'{col} - Distribution with IQR Bounds', fontweight='bold')
            axes[i, 1].legend()

            # Scatter plot with outliers highlighted
            y_pos = np.random.normal(0, 0.1, len(data))
            axes[i, 2].scatter(data, y_pos, alpha=0.6, color='blue', label='Normal')
            if len(iqr_outliers) > 0:
                outlier_y = np.random.normal(0, 0.1, len(iqr_outliers))
                axes[i, 2].scatter(iqr_outliers, outlier_y, alpha=0.8, color='red', label='IQR Outliers')
            axes[i, 2].set_title(f'{col} - Outlier Detection', fontweight='bold')
            axes[i, 2].set_xlabel(col)
            axes[i, 2].legend()
            axes[i, 2].set_yticks([])

    plt.tight_layout()
    plt.show()

    return pd.DataFrame(outlier_summary)

# Perform outlier analysis
if numerical_cols:
    print("="*60)
    print("COMPREHENSIVE OUTLIER ANALYSIS")
    print("="*60)
    outlier_results = comprehensive_outlier_analysis(df, numerical_cols)
    print(outlier_results)
```

```python

def create_correlation_analysis(df, target_col=None):
    """Create comprehensive correlation analysis"""
    numerical_df = df.select_dtypes(include=[np.number])

    if len(numerical_df.columns) < 2:
        print("Not enough numerical columns for correlation analysis")
        return

    # Correlation matrix
    corr_matrix = numerical_df.corr()

    fig, axes = plt.subplots(2, 2, figsize=(20, 16))

    # Correlation heatmap
    mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
    sns.heatmap(corr_matrix, mask=mask, annot=True, cmap='coolwarm', center=0,
                square=True, ax=axes[0, 0], cbar_kws={'shrink': 0.8})
    axes[0, 0].set_title('Correlation Matrix Heatmap', fontsize=14, fontweight='bold')

    # Clustermap
    sns.clustermap(corr_matrix, annot=True, cmap='coolwarm', center=0,
                   figsize=(10, 8), cbar_pos=(0.02, 0.83, 0.03, 0.15))
    plt.title('Hierarchical Clustering of Correlations')

    # Pairplot (sample if too many features)
    if len(numerical_df.columns) <= 6:
        if target_col and target_col in df.columns:
            sns.pairplot(df[numerical_df.columns.tolist() + [target_col]],
                        hue=target_col, diag_kind='kde', ax=axes[0, 1])
        else:
            sns.pairplot(numerical_df, diag_kind='kde')
        axes[0, 1].set_title('Pairplot of Numerical Features', fontsize=14, fontweight='bold')

    # Strong correlations
    strong_corr = []
    for i in range(len(corr_matrix.columns)):
        for j in range(i+1, len(corr_matrix.columns)):
            corr_val = corr_matrix.iloc[i, j]
            if abs(corr_val) > 0.5:  # Strong correlation threshold
                strong_corr.append({
                    'Feature_1': corr_matrix.columns[i],
                    'Feature_2': corr_matrix.columns[j],
                    'Correlation': corr_val
                })

    if strong_corr:
        strong_corr_df = pd.DataFrame(strong_corr).sort_values('Correlation', key=abs, ascending=False)
        print("STRONG CORRELATIONS (|r| > 0.5):")
        print(strong_corr_df)

    plt.tight_layout()
    plt.show()

    return corr_matrix

def analyze_categorical_relationships(df, categorical_cols, target_col=None):
    """Analyze relationships between categorical variables"""
    if len(categorical_cols) < 2:
        print("Not enough categorical columns for relationship analysis")
        return

    # Cram√©r's V matrix for categorical variables
    cramers_matrix = pd.DataFrame(index=categorical_cols, columns=categorical_cols)

    for col1 in categorical_cols:
        for col2 in categorical_cols:
            if col1 == col2:
                cramers_matrix.loc[col1, col2] = 1.0
            else:
                try:
                    cramers_matrix.loc[col1, col2] = cramers_v_stat(df[col1].dropna(), df[col2].dropna())
                except:
                    cramers_matrix.loc[col1, col2] = 0.0

    cramers_matrix = cramers_matrix.astype(float)

    # Visualize Cram√©r's V matrix
    plt.figure(figsize=(10, 8))
    mask = np.triu(np.ones_like(cramers_matrix, dtype=bool))
    sns.heatmap(cramers_matrix, mask=mask, annot=True, cmap='viridis',
                square=True, cbar_kws={'shrink': 0.8})
    plt.title("Cram√©r's V - Categorical Variable Associations", fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()

    return cramers_matrix

# Perform correlation analysis
target_column = 'Survived' if 'Survived' in df.columns else None
print("="*60)
print("CORRELATION ANALYSIS")
print("="*60)
correlation_matrix = create_correlation_analysis(df, target_column)

# Analyze categorical relationships
if len(categorical_cols) >= 2:
    print("\n" + "="*60)
    print("CATEGORICAL VARIABLE RELATIONSHIPS")
    print("="*60)
    cramers_results = analyze_categorical_relationships(df, categorical_cols, target_column)
```

```python

def create_engineered_features(df):
    """Create new features based on domain knowledge"""
    df_engineered = df.copy()

    # Age groups (if Age column exists)
    if 'Age' in df.columns:
        df_engineered['Age_Group'] = pd.cut(df['Age'],
                                          bins=[0, 12, 18, 35, 60, 100],
                                          labels=['Child', 'Teen', 'Adult', 'Middle_Age', 'Senior'])

    # Family size (if SibSp and Parch exist)
    if 'SibSp' in df.columns and 'Parch' in df.columns:
        df_engineered['Family_Size'] = df['SibSp'] + df['Parch'] + 1
        df_engineered['Is_Alone'] = (df_engineered['Family_Size'] == 1).astype(int)

    # Title extraction from Name (if Name column exists)
    if 'Name' in df.columns:
        df_engineered['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
        # Group rare titles
        title_counts = df_engineered['Title'].value_counts()
        rare_titles = title_counts[title_counts < 10].index
        df_engineered['Title'] = df_engineered['Title'].replace(rare_titles, 'Rare')

    # Cabin deck (if Cabin column exists)
    if 'Cabin' in df.columns:
        df_engineered['Deck'] = df['Cabin'].str[0]
        df_engineered['Has_Cabin'] = df['Cabin'].notna().astype(int)

    # Fare per person (if Fare and family size exist)
    if 'Fare' in df.columns and 'Family_Size' in df_engineered.columns:
        df_engineered['Fare_Per_Person'] = df['Fare'] / df_engineered['Family_Size']

    return df_engineered

# Create engineered features
df_engineered = create_engineered_features(df)

# Display new features
new_features = [col for col in df_engineered.columns if col not in df.columns]
if new_features:
    print("="*60)
    print("NEWLY ENGINEERED FEATURES")
    print("="*60)
    for feature in new_features:
        print(f"- {feature}")
        if df_engineered[feature].dtype == 'object':
            print(f"  Value counts:\n{df_engineered[feature].value_counts()}")
        else:
            print(f"  Statistics:\n{df_engineered[feature].describe()}")
        print()

# Analyze impact of new features on target variable
if target_column and target_column in df_engineered.columns:
    print("="*60)
    print("FEATURE IMPACT ON TARGET VARIABLE")
    print("="*60)

    for feature in new_features:
        if feature in df_engineered.columns:
            print(f"\n{feature} vs {target_column}:")
            crosstab = pd.crosstab(df_engineered[feature], df_engineered[target_column], normalize='index') * 100
            print(crosstab.round(2))
```

```python

def create_interactive_dashboard(df, target_col=None):
    """Create interactive Plotly visualizations"""

    # Interactive correlation heatmap
    numerical_df = df.select_dtypes(include=[np.number])
    if len(numerical_df.columns) > 1:
        corr_matrix = numerical_df.corr()

        fig_corr = go.Figure(data=go.Heatmap(
            z=corr_matrix.values,
            x=corr_matrix.columns,
            y=corr_matrix.columns,
            colorscale='RdBu',
            zmid=0,
            text=corr_matrix.round(3).values,
            texttemplate='%{text}',
            textfont={"size": 10},
            hoverongaps=False
        ))

        fig_corr.update_layout(
            title='Interactive Correlation Matrix',
            width=800,
            height=600
        )
        fig_corr.show()

    # Interactive scatter plot matrix
    if len(numerical_df.columns) >= 3:
        fig_scatter = px.scatter_matrix(
            numerical_df,
            color=df[target_col] if target_col and target_col in df.columns else None,
            title="Interactive Scatter Plot Matrix",
            height=800
        )
        fig_scatter.show()

    # Interactive box plots for numerical features
    for col in numerical_df.columns:
        if target_col and target_col in df.columns:
            fig_box = px.box(
                df,
                x=target_col,
                y=col,
                title=f'Interactive Box Plot: {col} by {target_col}',
                color=target_col
            )
        else:
            fig_box = px.box(
                df,
                y=col,
                title=f'Interactive Box Plot: {col}'
            )
        fig_box.show()

    # Interactive bar plots for categorical features
    categorical_df = df.select_dtypes(include=['object'])
    for col in categorical_df.columns:
        if col != target_col:
            if target_col and target_col in df.columns:
                fig_bar = px.histogram(
                    df,
                    x=col,
                    color=target_col,
                    title=f'Interactive Bar Plot: {col} by {target_col}',
                    barmode='group'
                )
            else:
                fig_bar = px.histogram(
                    df,
                    x=col,
                    title=f'Interactive Bar Plot: {col}'
                )
            fig_bar.show()

# Create interactive dashboard
print("="*60)
print("INTERACTIVE VISUALIZATIONS")
print("="*60)
create_interactive_dashboard(df_engineered, target_column)
```

```python

def generate_eda_report(df, df_engineered, target_col=None):
    """Generate comprehensive EDA summary report"""

    report = {
        'dataset_overview': {},
        'data_quality': {},
        'feature_insights': {},
        'relationships': {},
        'recommendations': []
    }

    # Dataset Overview
    report['dataset_overview'] = {
        'shape': df.shape,
        'memory_usage_mb': df.memory_usage(deep=True).sum() / 1024**2,
        'numerical_features': len(df.select_dtypes(include=[np.number]).columns),
        'categorical_features': len(df.select_dtypes(include=['object']).columns),
        'total_missing_values': df.isnull().sum().sum(),
        'missing_percentage': (df.isnull().sum().sum() / (df.shape[0] * df.shape[1])) * 100
    }

    # Data Quality Assessment
    missing_analysis = analyze_missing_values(df)
    report['data_quality'] = {
        'columns_with_missing': len(missing_analysis),
        'highest_missing_column': missing_analysis.iloc[0]['Column'] if len(missing_analysis) > 0 else None,
        'highest_missing_percentage': missing_analysis.iloc[0]['Missing_Percentage'] if len(missing_analysis) > 0 else 0
    }

    # Feature Insights
    numerical_cols = df.select_dtypes(include=[np.number]).columns
    categorical_cols = df.select_dtypes(include=['object']).columns

    if len(numerical_cols) > 0:
        skewed_features = []
        for col in numerical_cols:
            skewness = df[col].skew()
            if abs(skewness) > 1:
                skewed_features.append((col, skewness))

        report['feature_insights']['highly_skewed_features'] = skewed_features

    # Relationships
    if target_col and target_col in df.columns:
        numerical_df = df.select_dtypes(include=[np.number])
        if target_col in numerical_df.columns and len(numerical_df.columns) > 1:
            correlations_with_target = numerical_df.corr()[target_col].abs().sort_values(ascending=False)
            correlations_with_target = correlations_with_target[correlations_with_target.index != target_col]

            report['relationships']['top_correlated_features'] = [
                (feature, corr) for feature, corr in correlations_with_target.head(3).items()
            ]

    # Generate Recommendations
    recommendations = []

    if report['data_quality']['highest_missing_percentage'] > 50:
        recommendations.append(f"Consider dropping {report['data_quality']['highest_missing_column']} due to high missing values ({report['data_quality']['highest_missing_percentage']:.1f}%)")

    if len(report['feature_insights'].get('highly_skewed_features', [])) > 0:
        recommendations.append("Consider log transformation or other normalization techniques for highly skewed features")

    if report['dataset_overview']['missing_percentage'] > 10:
        recommendations.append("Implement comprehensive missing value imputation strategy")

    recommendations.append("Consider feature engineering based on domain knowledge")
    recommendations.append("Evaluate feature importance for model selection")

    report['recommendations'] = recommendations

    return report

# Generate and display report
eda_report = generate_eda_report(df, df_engineered, target_column)

print("="*80)
print("COMPREHENSIVE EDA SUMMARY REPORT")
print("="*80)

print("\n DATASET OVERVIEW")
print("-" * 40)
for key, value in eda_report['dataset_overview'].items():
    print(f"{key.replace('_', ' ').title()}: {value}")

print("\n DATA QUALITY ASSESSMENT")
print("-" * 40)
for key, value in eda_report['data_quality'].items():
    print(f"{key.replace('_', ' ').title()}: {value}")

print("\n KEY FEATURE INSIGHTS")
print("-" * 40)
if 'highly_skewed_features' in eda_report['feature_insights']:
    print("Highly Skewed Features:")
    for feature, skewness in eda_report['feature_insights']['highly_skewed_features']:
        print(f"  - {feature}: {skewness:.3f}")

print("\n RELATIONSHIPS")
print("-" * 40)
if 'top_correlated_features' in eda_report['relationships']:
    print("Top Features Correlated with Target:")
    for feature, corr in eda_report['relationships']['top_correlated_features']:
        print(f"  - {feature}: {corr:.3f}")

print("\n RECOMMENDATIONS")
print("-" * 40)
for i, recommendation in enumerate(eda_report['recommendations'], 1):
    print(f"{i}. {recommendation}")

# Save processed dataset
df_engineered.to_csv('processed_titanic_dataset.csv', index=False)
print(f"\n Processed dataset saved as 'processed_titanic_dataset.csv'")
print(f"Original shape: {df.shape}")
print(f"Processed shape: {df_engineered.shape}")
print(f"New features added: {len(df_engineered.columns) - len(df.columns)}")
```
